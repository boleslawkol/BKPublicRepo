# Analyzing and Interpreting Sales Data Using a Line Plot and Histogram
**Dataset Description:** You have a dataset containing monthly sales data for a tech company over 5 years. The dataset has two columns: Month-Year (formatted as 'Jan-2018', 'Feb-2018', etc.) and Sales (number of units sold).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
dates = pd.date_range("20180101", periods=60, freq='M')
sales = np.random.randint(50, 200, size=60) + np.linspace(0, 100, 60)

df = pd.DataFrame({'Month-Year': dates, 'Sales': sales})

# Line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month-Year', y='Sales', data=df)
plt.title('Monthly Sales Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Sales'], bins=15, kde=True)
plt.title('Distribution of Monthly Sales')
plt.show()
```

```python
sns.boxplot(x=df['Sales'])
```

# Scenario: Urban Transportation Management

**Problem:**
A city's urban transportation department wants to improve bus services. They've identified several problems:

1. Some buses are overcrowded while others are underutilized.
2. Delays in bus arrivals cause inconvenience.
3. Certain routes have frequent stops due to technical issues, causing further delays.

To address these, they've collected data and aim to model different aspects using various probability distributions.

## 1. Binomial Distribution:
**Problem:** Determine the likelihood of a particular bus being full on any given trip.

***Assumptions:*** Each bus trip has a 0.7 probability of being full. In 10 trips, how likely is it that exactly 5 trips are full?

```python
from scipy.stats import binom

# Parameters
n = 10  # number of trials
p = 0.7  # probability of success

# Probability of 5 successes in 10 trials
prob = binom.pmf(5, n, p)
print(prob)
```

- ***Technical explanation***
- ***Non-technical explanation***

## 2. Poisson Distribution:
**Problem:** Predict the number of buses arriving at a stop within a given hour.

***Assumptions:*** Buses arrive at a mean rate of 3 buses per hour. What's the probability of 5 buses arriving in the next hour?

```python
from scipy.stats import poisson

# Parameters
mu = 3  # mean

# Probability of 5 buses in an hour
prob = poisson.pmf(5, mu)
print(prob)
```

**Insights**

- ***Technical explanation***
- ***Non-technical explanation***

## 3. Geometric Distribution:
**Problem:** Determine how long, on average, riders have to wait for a non-full bus.

***Assumptions:*** The probability of a bus being non-full is 0.3. What's the expected waiting time for a non-full bus?

```python
from scipy.stats import geom

# Parameters
p = 0.3  # probability of success (finding a non-full bus)

# Expected waiting time
expected_time = geom.mean(p)
print(expected_time)
```

**Insights**

- ***Technical explanation***
- ***Non-technical explanation***

## 4. Negative Binomial Distribution:
**Problem:** Predict the number of full buses a rider will encounter before finding the third non-full bus.

***Assumptions:*** The probability of a bus being non-full is 0.3.

```python
from scipy.stats import nbinom

# Parameters
r = 3  # number of non-full buses
p = 0.3  # probability of finding a non-full bus

# Expected number of full buses before finding the third non-full bus
expected_full_buses = nbinom.mean(r, p)
print(expected_full_buses)
```

**Insights**

- ***Technical explanation***
- ***Non-technical explanation***

## 5. Normal Distribution:
**Problem:** Predict the distribution of time intervals between bus arrivals.

***Assumptions:*** Buses arrive every 20 minutes on average with a standard deviation of 5 minutes. What's the probability that the next bus will arrive in 15 to 25 minutes?

```python
from scipy.stats import norm

# Parameters
mu = 20  # mean
sigma = 5  # standard deviation

# Probability of the bus arriving between 15 to 25 minutes
prob = norm.cdf(25, mu, sigma) - norm.cdf(15, mu, sigma)
print(prob)
```

**Insights**

- ***Technical explanation***
- ***Non-technical explanation***

## 6. Exponential Distribution:
**Problem:** Model the time between bus breakdowns.

***Assumptions:*** Breakdowns occur, on average, every 50 hours. What's the probability that the next breakdown will happen before 30 hours?

```python
from scipy.stats import expon

# Parameters
scale = 50  # mean, or "scale" parameter for the exponential distribution

# Probability of a breakdown occurring before 30 hours
prob = expon.cdf(30, scale=scale)
print(prob)
```

# Exercise 1: Evaluating a New Product Launch

**Business Problem:** A company, "TechGlo", has recently launched a new product in three different cities: New York, Los Angeles, and Chicago. They want to understand if the sales patterns across these cities are significantly different, and if a specific promotional strategy employed in one city had an effect on sales. They also want to know if there's a relationship between the type of customers (categorized by age) and the product's popularity.




**Data:**

- Sales data from each city for the first month since the product launch.
Customer age data for each sale.
- Information on the promotional strategy used in each city.

**Charts:** Plot the sales data from each city on a bar chart

```python
import matplotlib.pyplot as plt
import numpy as np

cities = ['New York', 'Los Angeles', 'Chicago']
sales = [1200, 1350, 1100]  # Example data

plt.bar(cities, sales)
plt.title('Sales by City')
plt.xlabel('City')
plt.ylabel('Number of Sales')
plt.show()
```

***Insight:***

**ANOVA (One-way ANOVA):** Test if the mean sales across the three cities are significantly different.

```python
import scipy.stats as stats

# Sales data for New York
sales_ny = [120, 115, 130, 125, 123, 127, 128, 122, 119, 121,
            130, 129, 132, 133, 131, 134, 128, 126, 125, 123,
            121, 119, 120, 123, 127, 129, 130, 128, 124, 122]

# Sales data for Los Angeles
sales_la = [135, 138, 137, 136, 139, 140, 141, 142, 144, 145,
            143, 142, 141, 140, 138, 137, 135, 134, 133, 132,
            131, 130, 128, 127, 125, 126, 127, 128, 129, 130]

# Sales data for Chicago
sales_chi = [110, 112, 111, 113, 114, 113, 112, 110, 109, 108,
             107, 105, 104, 103, 102, 101, 100, 99, 98, 97,
             96, 95, 94, 93, 92, 90, 91, 90, 89, 88]

# Perform the ANOVA test as described in the exercise
import scipy.stats as stats
f_val, p_val = stats.f_oneway(sales_ny, sales_la, sales_chi)

print(f"Test statistic: {f_val}")
print(f"p-value: {p_val}")
```

***Insight:***

**Chi-Square Tests (Test of Independence):** Check if there's a relationship between the city and the promotional strategy employed.

```python
from scipy.stats import chi2_contingency

# Observed data in a (3x3 table) for 3 promotional strategies across 3 cities
observed = [
    [50, 40, 30],
    [30, 60, 40],
    [20, 30, 60]
]

chi2, p, _, _ = chi2_contingency(observed)
print(f"Test statistic: {chi2}")
print(f"p-value: {p}")
```

***Insight:***

**Probability Distributions (Discrete):** Considering the sales data follows a Poisson distribution, calculate the probability that on a given day, there are more than 100 sales in New York.

```python
mu = sum(sales_ny) / len(sales_ny)  # mean sales for New York

prob = 1 - stats.poisson.cdf(100, mu)
print(f"Probability: {prob}")
```

***Insight:***

**Sampling (Random Sampling, Sampling Distribution):** Take a random sample of 30 sales from each city and calculate the mean. Use these samples to describe the sampling distribution of the sample mean.

```python
sample_ny = np.random.choice(sales_ny, 30)
sample_mean_ny = np.mean(sample_ny)

print(f"Sample mean NY: {sample_mean_ny}")
```

***Insight:***

**Hypothesis Testing (Two-sample tests):** Test if the mean sales between New York and Los Angeles are significantly different.

```python
t_val, p_val_two_sample = stats.ttest_ind(sales_ny, sales_la)

print(f"Test statistic: {t_val}")
print(f"p-value: {p_val_two_sample}")
```
