```
webpage = 'https://www.simplilearn.com/'

try:
    html = urlopen(webpage)
except HTTPError as e:
    print("The server returned an HTTP error")
except URLError as e:
    print("The server could not be found!")
else:
    print(html.read())
```

```
url = 'https://www.imdb.com/search/title?release_date=2018&sort=boxoffice_gross_us,desc&start=1'
movies_req = urlopen(url)
bs = BeautifulSoup(movies_req, 'html.parser')
```

```
movie_blocks = bs.findAll('div',{'class':'lister-item-content'})
```

```
m_reyear = movie_blocks[0].find('span',{'class': 'lister-item-year'}).get_text()
m_reyear
```
